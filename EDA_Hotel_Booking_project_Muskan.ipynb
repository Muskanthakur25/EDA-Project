## **Handling Missing values **

````python
hotel_booking_df.isnull().sum() #it is providing us the number of missing values in each column
# here we found that column(children =4 null values) (country =452) and (agent =12193) null values
````

````python
hotel_booking_df['children'].fillna(0, inplace=True)# here we are filling the null values with 0.Presuming there are no children in the family
````

````python
# similarly the agent column has 12193 null values which can cause issue later in data insights(e.g while aggregating and grouping it may lead to incorrect results)
# So we can fill these null values with 0 too.
hotel_booking_df['agent'].fillna(0,inplace=True)
````

````python
#For country column we can handle its null values with "Unknown" or "Not Available" either.

hotel_booking_df['country'].fillna('Unknown', inplace=True)
````

````python
# check if any null values are still present in the dataset
hotel_booking_df.isnull().sum()
````

**Data Transformation(changing inappropriate datatype of attributes whereever required)**

````python
# first we'll find the datatypes of all the columns
hotel_booking_df.info()
````

````python
# We've observed from above that the agent and children are of the float type, which is not required.
# Convert multiple columns from float to int
columns_to_convert = ['children', 'agent']  # Replace with your column names
hotel_booking_df[columns_to_convert] = hotel_booking_df[columns_to_convert].apply(lambda x: x.astype('int64'))
````

**Data Integration (by joining columns)**

````python
# Calculate the cancellation rate for each room type
hotel_booking_df['cancellation_rate'] = hotel_booking_df['is_canceled'] / hotel_booking_df['stays_in_week_nights']

````

````python
# Create a new column for the total duration of stay
hotel_booking_df['total_stay_nights'] = hotel_booking_df['stays_in_weekend_nights'] + hotel_booking_df['stays_in_week_nights']
````

````python
hotel_booking_df['total_price'] = hotel_booking_df['adr'] * hotel_booking_df['total_stay_nights']# to calculate revenue only for confirmed bookings or specific periods.

````

````python
# Filter the DataFrame for the specific hotel (e.g., "city hotel")
specific_hotel_data = hotel_booking_df[hotel_booking_df['hotel'] == 'City Hotel']

# Calculate the total revenue by summing the 'total_price' column
total_revenue = specific_hotel_data['total_price'].sum()

print(f"Total Revenue for City Hotel: {total_revenue}")
````

````python
hotel_booking_df.tail(40)
````

````python
# # Create a dictionary mapping countries to regions
country_to_region = {
    'USA': 'North America',
    'Germany': 'Europe',
    'India': 'Asia',
    'France': 'Europe',
    'Japan': 'Asia',
    'Brazil': 'South America',
    'Australia': 'Oceania',
    # Add more countries as needed
}

# Apply this mapping to create a 'guest_nationality' column
hotel_booking_df['guest_nationality'] = hotel_booking_df['country'].map(country_to_region)

# If a country is not in the dictionary, it will be NaN, so you may want to fill those with a default region (e.g., 'Other')
hotel_booking_df['guest_nationality'].fillna('Other')
````

````python
# Create an 'age_group' based on the number of children, adults, and babies
def age_group(num):
    if num['children'] > 0:
        return 'Family'
    elif num['adults'] >= 2:
        return 'Couple'
    else:
        return 'Solo'

hotel_booking_df['guest_type'] = hotel_booking_df.apply(age_group, axis=1)
````

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code

corr = hotel_booking_df[hotel_booking_df.describe().columns].corr()
cmap = sns.diverging_palette(220, 20, as_cmap=True)
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.2)
sns.heatmap(corr, cmap=cmap, annot=False, fmt=".2f", annot_kws={"size": 12, "weight": "bold"},
            linewidths=0.5, cbar_kws={"shrink": 0.8, "label": "Correlation Coefficient"},
            square=True, linecolor="white", vmin=-1, vmax=1)
plt.title("Correlation Heatmap", fontsize=18)
plt.yticks(rotation=0)
# Show the plot
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer Here. Identifying Relationships: It helps identify which variables are positively correlated, negatively correlated, or not correlated at all. Positive correlations (close to +1) suggest that as one variable increases, the other tends to increase, while negative correlations (close to -1) indicate that as one variable increases, the other tends to decrease.

##### 2. What is/are the insight(s) found from the chart?

Answer Here Strong positive or negative correlations can be shown in a heatmap.
Positive correlations could point to cross-selling, bundling, or shared elements that affect company success.

High correlations between variables may be a sign of overlapping or redundant data.
Finding these facilitates data collecting and boosts operational effectiveness.

Opportunities to enhance one variable by concentrating on another may be indicated by negative correlations.
